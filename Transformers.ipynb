```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from sentence_transformers import SentenceTransformer

from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('/content/drive/MyDrive/datasets/titles.csv')
print(df.shape)
df.head()
```

```python
sample = df.sample(50)
sample = sample.reset_index(drop=True)
sample.head()
```

```python
sample
```

```python
#cleaning
sample['text'] = sample['title'].str.lower()
sample['text'] = sample['text'].apply(lambda item: " ".join(item for item in item.split() if item not in stop))
sample.head()
```

```python
model = SentenceTransformer('sentence-transformers/multi-qa-MiniLM-L6-cos-v1')
```

```python
print(model)
```

```python
features = model.encode(sample['text'])
sample['features'] = list(features)
sample.head()
```

```python
def search(query_text):
   query_feature = model.encode([query_text])
   sample['similarity'] = sample['features'].apply(lambda x: cosine_similarity([x], query_feature)[0][0])
   sorted = sample.sort_values(by='similarity', ascending=False)
   return sorted.head(10)
```

```python
search('deep learning')
```

```python
def results(query_text):
   query_feature = model.encode([query_text])
   sample['similarity'] = sample['features'].apply(lambda x: cosine_similarity([x], query_feature)[0][0])
   sorted = sample.sort_values(by='similarity', ascending=False)
   result = sorted['title'].head(5)
   return list(result)
```

```python
results("Maths")
```

```python
import gradio as gr

u_interface = gr.Interface(fn=results, inputs="text", outputs="text", debug=False)
u_interface.launch(share=True)
```
